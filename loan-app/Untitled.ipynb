# CLUSTERING OF DATA

```python
approved = data[data['loan_status'] == 'Approved']
approved
```

```python
rejected = data[data['loan_status'] == 'Rejected']
rejected
```

```python
len(approved) + len(rejected) == len(data)
```

```python
data.loan_status.unique()
```

```python
pyp.yticks([])
pyp.xticks([])
pyp.tick_params(bottom = False, left = False)
ax = pyp.gca()
for location in ["left","right","bottom","top"]:
    ax.spines[location].set_visible(False)

pyp.bar(data.loan_status.unique(),[len(approved),len(rejected)],width=0.35,\
        color=['brown','gray'],alpha=1)
ax.set_xticks(data.loan_status.unique(),weight='bold')
pyp.ylabel('Success Ratio', fontweight='bold')
ax.text(-0.1,2680,str((round(len(approved)/len(data)*100,2)))+'%',weight='bold')
ax.text(0.9,1680,str((round(len(rejected)/len(data)*100,2)))+'%',weight='bold')
ax.text(0.4,2280,'LOAN STATUS',weight='bold')
pyp.show()
```

```python
avg = approved[(approved['cibil_score'] >= 500) & (approved['cibil_score'] <= 649)]
avg
```

```python
gud = approved[(approved['cibil_score'] >= 650) & (approved['cibil_score'] <= 749)]
gud
```

```python
r_gud = rejected[(rejected['cibil_score'] >= 650) & (rejected['cibil_score'] <= 749)]
r_gud
```

# VISUALISATION AND LOGICAL CONDITIONS

```python
pyp.xlim(-0.5, 0.8)
pyp.scatter(exc['loan_status'],exc['cibil_score'],color='blue',s=50)
pyp.scatter(gud['loan_status'],gud['cibil_score'],color='green',s=50)
pyp.scatter(avg['loan_status'],avg['cibil_score'],color='orange',s=50)
pyp.scatter(poor['loan_status'],poor['cibil_score'],color='red',s=50)
ax = pyp.gca()

ax.text(0.1,850,str(round((len(exc)/len(approved)*100),2) )+ '%',weight='bold')
ax.text(0.1,700,str(round((len(gud)/len(approved)*100),2) )+ '%',weight='bold')
ax.text(0.1,600,str(round((len(avg)/len(approved)*100),2) )+ '%',weight='bold')
ax.text(0.1,400,str(round((len(poor)/len(approved)*100),2) )+ '%',weight='bold')

pyp.ylabel('Credit Score',weight='bold')

pyp.show()
```

```python
pyp.xlim(-0.5, 0.8)
pyp.scatter(r_exc['loan_status'],r_exc['cibil_score'],color='blue',s=50)
pyp.scatter(r_gud['loan_status'],r_gud['cibil_score'],color='green',s=50)
pyp.scatter(r_avg['loan_status'],r_avg['cibil_score'],color='orange',s=50)
pyp.scatter(r_poor['loan_status'],r_poor['cibil_score'],color='red',s=50)
ax = pyp.gca()

ax.text(0.1,850,str(round((len(r_exc)/len(rejected)*100),2) )+ '%',weight='bold')
ax.text(0.1,700,str(round((len(r_gud)/len(rejected)*100),2) )+ '%',weight='bold')
ax.text(0.1,600,str(round((len(r_avg)/len(rejected)*100),2) )+ '%',weight='bold')
ax.text(0.1,400,str(round((len(r_poor)/len(rejected)*100),2) )+ '%',weight='bold')

pyp.ylabel('Credit Score',weight='bold')

pyp.show()
```
