# DATA WRANGLING

```python
empty = data[(data.START == 'Unknown Location') | (data.STOP == 'Unknown Location')]
empty
```

```python
len(empty)
```

```python
full_empty = data[(data.START == 'Unknown Location') & (data.STOP == 'Unknown Location')]
full_empty.head()
```

```python
len(full_empty)
```

```python
full_empty.index
```

```python
for indexes in full_empty.index:
    data.drop(indexes,inplace=True)
```

```python
data.PURPOSE.value_counts()
```

```python
data.PURPOSE.isnull().sum()
```

```python
purpose_empty = data[data.PURPOSE.isnull()]
purpose_empty.head()
```

```python
purpose_empty.index
```

```python
len(data)
```

```python
for indexes in purpose_empty.index:
    data.loc[indexes, 'PURPOSE'] = 'Entertainment'
```

```python
data.PURPOSE.isnull().sum()
```

```python
data.PURPOSE.value_counts()
```

```python
partial_empty = data[(data.START == 'Unknown Location') | (data.STOP == 'Unknown Location')]
partial_empty
```

```python
partial_empty.index
```

```python
for index in partial_empty.index:
    temp = data.loc[index, 'START']
    if temp != 'Unknown Location':
        data.loc[index, 'STOP'] = temp
    else:
        temp = data.loc[index, 'STOP']
        data.loc[index, 'START'] = temp
```

```python
data.tail(20)
```

```python
data['DURATION'] = (data.END_DATE-data.START_DATE).astype('timedelta64[m]')
```

```python
data.head()
```

# VISUALISATION OF INSIGHTS

```python
fig,ax = pyp.subplots(2,1,figsize=(9,8))

ax[0].tick_params(bottom = False, left = False),
ax[1].tick_params(bottom = False,left = False)

for location in ["left","right","bottom","top"]:
    ax[0].spines[location].set_visible(False)
    ax[1].spines[location].set_visible(False)   

ax[0].plot(bus_dur.DURATION.value_counts().sort_index(),color='red')
ax[1].plot(per_dur.DURATION.value_counts().sort_index(),color='darkblue')

ax[0].text(150,40,'NUMBER OF FREQUENCY AGAINST',weight='bold')
ax[0].text(160,35,'DURATION OF UBER TRIP IN MINUTES',weight='bold')
    
ax[1].text(90,5,'NUMBER OF FREQUENCY AGAINST',weight='bold')
ax[1].text(100,4.5,'DURATION OF UBER TRIP IN MINUTES',weight='bold')
pyp.show()
```

```python
temp = data[data.PURPOSE == 'Entertainment']
int(sum(temp.MILES))
```

```python
x,y = [],[]
for indexes in data.PURPOSE.value_counts().index:
    temp = data[data.PURPOSE == indexes]
    s = int(sum(temp.MILES))
    x.append(indexes)
    y.append(s)
```

```python
x,y
```

```python
fig,ax = pyp.subplots(1,1,figsize=(8,7))

ax.tick_params(bottom = False, left = False)

for location in ["left","right","bottom","top"]:
    ax.spines[location].set_visible(False)  

colors=['palegreen','skyblue','slategrey','grey','royalblue','navy']
ax.bar(x,y,color=colors)
ax.set_ylabel('Mileage Coverage', fontweight='bold')
ax.set_xticklabels(x,rotation=90)

ax.text(2,4500,'MILEAGE COVERAGE AGAINST TRIPS',weight='bold')
pyp.show()
```
