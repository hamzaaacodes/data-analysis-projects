```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyp
import seaborn as sns
file = pd.read_csv(r"C:\Users\Hp\Desktop\olx_car_data_csv.csv")
y_1 = file["Year"]
p_1 = file["Price"]
c_1 = file["Registered City"]
C = file["Fuel"]
dataframe = ({"Car_Price":p_1,"Model_Year":y_1})
pyp.xlabel('Model_Year')
pyp.ylabel('Car_Price')
pyp.ylim(ymin=100000,ymax=4000000)
pyp.ticklabel_format(style='plain', axis='y')
a = sns.scatterplot(x=y_1,y=p_1,data=dataframe,hue=C)
a.set(xlim=(1980,2019))
pyp.xticks(range(1980,2019,1),rotation=90,ha='center')
pyp.tight_layout()
pyp.plot(file.groupby(['Year']).Price.mean())
pyp.show()
```
