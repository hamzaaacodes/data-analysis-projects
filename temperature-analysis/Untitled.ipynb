# IMPORTING LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as pyp
```

# DATA READING

```python
data = pd.read_csv('Tempreture_1901_2016_Pakistan.csv')
data
```

# DATA ANALYSING

```python
data.columns
```

```python
data.columns = data.columns.str.lstrip()
data.columns
```

```python
data = data[data['Year'] >= 2007]
data
```

```python
data = data.reset_index(drop=True)
data
```

```python
data.head()
```

```python
data.rename(columns={'Temperature - (Celsius)': 'Temperature'},inplace=True)
```

# FETCHING 10 YEARS OF DATA

```python
yr = 2007
x,y = [], []
for i in range(10):
    data_new = data[(data['Year'] >= yr) & (data['Year'] < (yr+1))]
    val = round(data_new['Temperature'].mean(),2)
    x.append(yr),y.append(val)
    yr += 1
```

```python
data
```

```python
data['Temperature'] = round(data['Temperature'],2)
```

# VISUALISATION AND LOGICS

```python
data
```

```python
pyp.ylim(15,25)
pyp.plot(x,y,color='red',\
             marker='*',markerfacecolor='black',markersize=10)
pyp.xticks(x,rotation=90)
ax = pyp.gca()
ax.text(2007,23,'AVERAGE TEMPERATURE OF PER YEAR FROM 2007 TO 2016',weight='bold')
pyp.show()
```

```python
c = ['purple','blue','red','orange','green','gray','aqua','brown','olive','royalblue']
yr = 2007
for i in range(10):
    data_new = data[data['Year'] == yr]
    data_new = data_new.reset_index(drop=True)
    pyp.ylim(6,32)
    pyp.plot(data_new['Month'],data_new['Temperature'].tolist(),color=c[i],\
             marker='.',markerfacecolor='black',markersize=10)
    pyp.xticks(rotation=90)
    ax = pyp.gca()
    ax.text('June',20,'Year Of ' + str(data_new.iloc[i,1]),weight='bold')
    pyp.show()
    yr += 1
```

# 10 YEARS TOGETHER TEMPERATURE VISUALISATION

```python
c = ['purple','blue','red','orange','green','gray','aqua','brown','olive','royalblue']
yr = 2007
for i in range(10):
    data_new = data[data['Year'] == yr]
    data_new = data_new.reset_index(drop=True)
    pyp.ylim(6,35)
    pyp.plot(data_new['Month'],data_new['Temperature'].tolist(),color=c[i],\
             marker='.',markerfacecolor='black',markersize=10)
    
    yr += 1
pyp.xticks(rotation=90)
ax = pyp.gca()
ax.text('March',33,'10 YEARS TEMPERATURE MEASUREMENT',weight='bold')
pyp.show()
```
