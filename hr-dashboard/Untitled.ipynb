# DATA WRANGLING

```python
for i in range(len(data['Birth_Date'])):
    splitted = data.iloc[i,0].split('-')
    if len(splitted[-1]) == 2:
        data.iloc[i,0] = splitted[0] + '/' + splitted[1] + '/' + '19' + splitted[-1]
```

```python
for i in range(len(data['Hire_Date'])):
    splitted = data.iloc[i,-3].split('-')
    if len(splitted[-1]) == 2:
        data.iloc[i,-3] = splitted[0] + '/' + splitted[1] + '/' + '19' + splitted[-1]
```

```python
data.head()
```

```python
data['Age'] = ''
```

```python
for i in range(len(data['Birth_Date'])):
    least = data.iloc[i,0]
    least = least[-4:]
    print(least)
```

```python
for i in range(len(data['Birth_Date'])):
    least = data.iloc[i,0]
    least = least[-4:]
    data['Age'][i] = 2023 - int(least)
```

```python
data.head(50)
```

```python
data['Age_Category'] = ''
```

```python
for i in range(len(data['Age'])):
    if data['Age'][i] <= 35:
        data['Age_Category'][i] = 'Young'
    elif (data['Age'][i] > 35) and (data['Age'][i] <= 55):
        data['Age_Category'][i] = 'Old'
    else:
        data['Age_Category'][i] = 'Very Old'
```
